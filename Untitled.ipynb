```python
import pandas as pd 
import numpy as np 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import difflib
```

```python
movies_data=pd.read_csv('movies.csv')
movies_data.head()
```

```python
 selected_features=['genres','keywords','tagline','cast','director']
for features in selected_features:
    movies_data[features]=movies_data[features].fillna('')
```

```python
movies_data['combined_features']=movies_data['genres']+''+movies_data['keywords']+''+movies_data['tagline']+''+movies_data['cast']+''+movies_data['director']
```

```python
movies_data['combined_features'].head()
```

```python
vectorizer=TfidfVectorizer()
vectorized_feature=vectorizer.fit_transform(movies_data['combined_features'])
vectorized_feature
```

```python
similarity=cosine_similarity(vectorized_feature)
```

```python
similarity
```

```python
input_data=input('Enter the movie name:')
title_list=movies_data['title'].tolist()
diff=difflib.get_close_matches(input_data,title_list)
print(diff)
```

```python
movie=diff[0]
```

```python
index_of_movie=movies_data[movies_data.title==movie]['index'].values[0]
print(index_of_movie)
```

```python
founded_similarity=list(enumerate(similarity[index_of_movie]))
```

```python
founded_similarity
```

```python
sorted_similarity=sorted(founded_similarity,key=lambda x:x[1],reverse=True)
sorted_similarity
```

```python
i=1
for movies in sorted_similarity:
    index=movies[0]
    title=movies_data[movies_data.index==index]['title'].values[0]
    if i<11:
        print(i,'.',title)
        i=i+1
  
    
    
         
   
    
    


    
```
